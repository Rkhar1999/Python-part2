# What is Statistics?

**Statistics** is the branch of mathematics that deals with the collection, analysis, interpretation, presentation, and organization of data. It provides tools and methods to understand data, identify patterns, and make informed decisions. Statistics is widely used in various fields such as business, medicine, social sciences, and technology to analyze real-world phenomena.

### Types of Statistics:
1. **Descriptive Statistics**: 
   - Summarizes and describes the features of a dataset.
   - Examples: Mean, Median, Mode, Standard Deviation, and Visualizations like charts or graphs.

2. **Inferential Statistics**: 
   - Uses data from a sample to make predictions or generalizations about a population.
   - Examples: Hypothesis Testing, Confidence Intervals, Regression Analysis.

### **Population and Sample**

#### **1. What is Population?**
A **population** in statistics refers to the entire group that you are interested in studying or analyzing. It includes all individuals or items that meet specific criteria. The population can be finite (e.g., all students in a school) or infinite (e.g., all the stars in the universe).

**Example**:
- The population of a country.
- All employees in a company.
- Every car produced by a manufacturer in a year.

#### **2. What is a Sample?**
A **sample** is a subset of the population that is selected for the purpose of analysis. It is often impossible or impractical to study an entire population, so a sample is used to make inferences about the population.

**Key Properties**:
- A sample should be representative of the population.
- It is usually selected randomly to avoid bias.

**Example**:
- A survey of 100 students to understand the preferences of all students in a school.
- Testing 1,000 cars from a production line to infer quality control for the entire batch.

---

### **Population vs Sample**

| **Aspect**            | **Population**                               | **Sample**                                 |
|------------------------|----------------------------------------------|--------------------------------------------|
| **Definition**         | Entire group of interest                    | Subset of the population                   |

| **Analysis**           | Harder to study directly                    | Easier to study                            |
| **Purpose**            | Provides complete insight                   | Used to make inferences about the population |
| **Example**            | All students in a school                    | 50 randomly selected students              |

### **Parameters and Sample Statistics**

---

#### **1. What is a Parameter?**
A **parameter** is a numerical value that describes a characteristic of a **population**. Since the population includes every member of the group, parameters are often difficult or impossible to measure directly. They are fixed values but unknown in most real-world cases.

**Examples of Parameters**:
- The average age of all people in a country (population mean).
- The percentage of defective products in all items produced by a factory.

---

#### **2. What is a Sample Statistic?**
A **sample statistic** is a numerical value that describes a characteristic of a **sample** (a subset of the population). Sample statistics are used as estimates of population parameters. Since they are based on samples, they can vary depending on the sample selected.

**Examples of Sample Statistics**:
- The average age of 100 people randomly selected from a country.
- The percentage of defective products in a sample of 500 items.

---

#### **Key Difference: Parameter vs Sample Statistic**
| **Aspect**           | **Parameter**                                  | **Sample Statistic**                          |
|-----------------------|-----------------------------------------------|-----------------------------------------------|
| **Definition**        | Describes a characteristic of the population  | Describes a characteristic of a sample        |
| **Value**             | Fixed but often unknown                       | Varies depending on the sample                |
| **Examples**          | Population mean (\( \mu \)), population proportion (\( p \)) | Sample mean (\( \bar{x} \)), sample proportion (\( \hat{p} \)) |
| **Purpose**           | Describes the entire population               | Used to estimate population parameters        |

---

#### **Simple Example**
Imagine we want to calculate the **average age** of passengers on the Titanic.

- The **population parameter** is the actual average age of all passengers on the Titanic (denoted as \( \mu \)).
- The **sample statistic** is the average age calculated from a sample of passengers (denoted as \( \bar{x} \)).

---

# **Types of Data**

#### **1. Types of Data Based on Nature**

##### **1.1. Qualitative Data (Categorical Data)**
- Represents categories or groups.
- Non-numerical in nature, though numbers can sometimes be used as labels.
- **Examples**:
  - Gender: Male, Female
  - Eye Color: Blue, Green, Brown
  - Shirt Sizes: Small, Medium, Large

**Types of Qualitative Data**:
- **Nominal Data**: Categories with no inherent order. (e.g., Eye color)
- **Ordinal Data**: Categories with a meaningful order, but the intervals between them are not consistent. (e.g., Shirt sizes)

##### **1.2. Quantitative Data (Numerical Data)**
- Represents measurable quantities.
- Always numerical in nature.
- **Examples**:
  - Age: 25, 30, 35
  - Salary: $50,000, $60,000, $70,000

**Types of Quantitative Data**:
- **Discrete Data**: Counts or whole numbers. (e.g., Number of students in a class)
- **Continuous Data**: Measured values that can take any value within a range. (e.g., Height, Weight)

---

### **2. Types of Data Based on Measurement Levels**

#### **2.1. Nominal Data**
- Categories with no logical order.
- Example: Blood groups (A, B, AB, O).

#### **2.2. Ordinal Data**
- Categories with a logical order, but intervals are not equal.
- Example: Movie ratings (1 star, 2 stars, 3 stars).

#### **2.3. Interval Data**
- Numerical data with meaningful intervals, but no true zero.
- Example: Temperature in Celsius or Fahrenheit.

---

# **Measures of Central Tendency**

### **Measures of Central Tendency**

**Measures of central tendency** describe the central or typical value of a dataset. They summarize data with a single representative value and are used to understand the distribution of data.

---

### **1. Types of Measures of Central Tendency**

#### **1.1. Mean (Arithmetic Average)**
The mean is the sum of all values divided by the total number of values. It is the most commonly used measure but can be affected by extreme values (outliers).

**Formula**:
$$\[
\text{Mean} = \frac{\text{Sum of all values}}{\text{Total number of values}}
\]$$

**Example**: The scores [10, 20, 30] have a mean of:
\$$[
\text{Mean} = \frac{10 + 20 + 30}{3} = 20
\]$$

#### **1.2. Median**
The median is the middle value when the data is sorted in ascending or descending order. It is less affected by outliers and skewed data.

- If the number of observations is **odd**, the median is the middle value.
- If the number of observations is **even**, the median is the average of the two middle values.

**Example**:
- For [10, 20, 30], the median is 20.
- For [10, 20, 30, 40], the median is \$ \frac{20+30}{2} = 25 \$).

---

#### **1.3. Mode**
The mode is the most frequently occurring value in the dataset. A dataset can have one mode, multiple modes, or no mode.

**Example**:
- For [10, 20, 20, 30], the mode is 20.
- For [10, 10, 20, 20, 30], the modes are 10 and 20 (bimodal).

---

### **Comparison of the Measures**

| **Measure**  | **Best Use Case**                              | **Limitations**                               |
|--------------|------------------------------------------------|-----------------------------------------------|
| **Mean**     | Ideal for symmetrical data.                    | Sensitive to outliers and skewed data.        |
| **Median**   | Robust for skewed data or data with outliers.  | Ignores the magnitude of data values.         |
| **Mode**     | Best for categorical data or to find frequency.| May not exist or may not be unique.           |

---

````python
### **Python Code Example Using Pandas**


import pandas as pd

# Create a sample dataset
data = {'Scores': [50, 60, 70, 80, 90, 100, 100]}
df = pd.DataFrame(data)

# Calculate measures of central tendency
mean = df['Scores'].mean()  # Mean
median = df['Scores'].median()  # Median
mode = df['Scores'].mode()[0]  # Mode (selecting the first mode in case of multiple modes)
````

# **Outlier**


An **outlier** is a data point that is significantly different from the rest of the data. It lies far away from the central tendency of the dataset and can distort statistical analyses like the mean and standard deviation.

---

### **Why Are Outliers Important?**
1. **Impact on Analysis**:
   - Outliers can skew the results of statistical calculations like mean and standard deviation.
2. **Significance**:
   - An outlier may indicate variability in the data, measurement errors, or a unique case worth exploring.

---

### **Types of Outliers**
1. **Univariate Outliers**:
   - Outliers identified by analyzing a single variable.
   - Example: A height of 250 cm in a dataset where the average height is around 170 cm.

2. **Multivariate Outliers**:
   - Outliers identified by analyzing multiple variables simultaneously.
   - Example: A person with low income and very high expenditure.

---

### **Methods to Detect Outliers**

#### **1. Statistical Methods**
- **Z-Score**:
  - Measures how far a data point is from the mean in terms of standard deviations.
  - A Z-score greater than 3 or less than -3 is often considered an outlier.
  
  **Formula**:
  $$\[
  Z = \frac{(X - \mu)}{\sigma}
  \]
  Where:
  - \( X \) = Data point
  - \( \mu \) = Mean
  - \( \sigma \) = Standard deviation$$

- **Interquartile Range (IQR)**:
  - Detects outliers based on the spread of the middle 50% of the data.
  - Outliers lie below \$( Q1 - 1.5 \times \text{IQR} \) or above \( Q3 + 1.5 \times \text{IQR} \$).

  **Formula**:
  \$[
  \text{IQR} = Q3 - Q1
  \$]

#### **2. Visualization Methods**
- Box plots
- Scatter plots

---

### **Python Code Example Using Pandas**

````python
import pandas as pd
import numpy as np

# Sample dataset
data = {'Scores': [50, 60, 70, 80, 90, 100, 200]}  # 200 is an outlier
df = pd.DataFrame(data)

# Method 1: Detect outliers using IQR
Q1 = df['Scores'].quantile(0.25)  # First quartile (25th percentile)
Q3 = df['Scores'].quantile(0.75)  # Third quartile (75th percentile)
IQR = Q3 - Q1  # Interquartile range

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = df[(df['Scores'] < lower_bound) | (df['Scores'] > upper_bound)]

# Method 2: Detect outliers using Z-score
mean = df['Scores'].mean()
std_dev = df['Scores'].std()

df['Z-Score'] = (df['Scores'] - mean) / std_dev
outliers_z = df[abs(df['Z-Score']) > 3]
````

````python
# Display results
print("Dataset:")
print(df)

print("\nOutliers Detected Using IQR:")
print(outliers_iqr)

print("\nOutliers Detected Using Z-Score:")
print(outliers_z)
````

### **Key Insights**
1. **Outlier Detection**:
   - The value 200 is identified as an outlier using both IQR and Z-score methods.
2. **IQR**:
   - Focuses on the central 50% of the data.
3. **Z-Score**:
   - Measures deviation from the mean, useful for normally distributed data.

# **Measures of Dispersion**

**Measures of dispersion** describe the spread or variability in a dataset. They provide insight into how much the data points deviate from the central tendency (mean, median, or mode).

---

### **Why Are Measures of Dispersion Important?**
- They complement central tendency measures by showing how data is distributed.
- They help identify consistency and variability in the data.
- They are essential for understanding outliers and the overall data range.

---

### **Types of Measures of Dispersion**

#### **1. Range**
The difference between the maximum and minimum values in the dataset.

**Formula**:
\$[
\text{Range} = \text{Maximum Value} - \text{Minimum Value}
\$]

**Example**: For [10, 20, 30, 40, 50], the range is:
\$[
\text{Range} = 50 - 10 = 40
\$]

---

#### **2. Interquartile Range (IQR)**
The range of the middle 50% of the data, calculated as the difference between the 75th percentile (\( Q3 \)) and the 25th percentile (\( Q1 \)).

**Formula**:
\$[
\text{IQR} = Q3 - Q1
\$]

**Example**: For [10, 20, 30, 40, 50], the quartiles are:
- \$( Q1 = 20 \), \( Q3 = 40 \)
\[
\text{IQR} = 40 - 20 = 20
\$]

---

#### **3. Variance**
The average of the squared differences between each data point and the mean. It measures how far data points spread out around the mean.

**Formula**:
\$[
\text{Variance} (\sigma^2) = \frac{\sum (X - \mu)^2}{N}
\$]
Where:
- \$( X \$) = Data point
- \($ \mu \$) = Mean
- \($ N \$) = Number of data points

---

#### **4. Standard Deviation (SD)**
The square root of variance. It indicates the average amount by which data points differ from the mean.

**Formula**:
\[$
\text{SD} = \sqrt{\text{Variance}}
\]$

# **Empirical Rule (68-95-99.7 Rule)**

The **Empirical Rule** is a statistical principle that applies to **normal distributions** (bell-shaped curves). It states:

1. **68%** of the data falls within **1 standard deviation** of the mean.
2. **95%** of the data falls within **2 standard deviations** of the mean.
3. **99.7%** of the data falls within **3 standard deviations** of the mean.

This rule helps in understanding how data is distributed in a normal distribution.

---

### **Why Is the Empirical Rule Useful?**
- It provides a quick estimate of the spread of data.
- It helps identify **outliers** (data points beyond \( 3\sigma \)).
- It simplifies interpreting standard deviation in real-world problems.

---

### **Simple Example**

Imagine a class of students has scores on a test with:
- **Mean (\( \mu \)) = 70**
- **Standard Deviation (\( \sigma \)) = 10**

#### **Empirical Rule Application**
1. **68% Rule (1 SD)**:
   - \( \mu Â± 1\sigma = 70 Â± 10 \)
   - Range: [60, 80]
   - **68%** of students scored between 60 and 80.

2. **95% Rule (2 SDs)**:
   - \( \mu Â± 2\sigma = 70 Â± 20 \)
   - Range: [50, 90]
   - **95%** of students scored between 50 and 90.

3. **99.7% Rule (3 SDs)**:
   - \( \mu Â± 3\sigma = 70 Â± 30 \)
   - Range: [40, 100]
   - **99.7%** of students scored between 40 and 100.

Any score outside this range (below 40 or above 100) is a potential **outlier**.

---



### **Coefficient of Variation (CV)**

The **Coefficient of Variation (CV)** is a measure of relative variability. It expresses the standard deviation as a percentage of the mean, making it useful for comparing the degree of variation across datasets with different units or scales.

### **Formula**

\$[
\text{CV} = \left( \frac{\text{Standard Deviation}}{\text{Mean}} \right) \times 100
\$]

Where:
- **Standard Deviation (SD)** measures the spread of the data.
- **Mean** is the average value of the dataset.

---

### **Key Features**
- The **CV** is unitless, allowing comparisons across datasets with different units or scales.
- It is typically used when comparing variability between datasets or measuring consistency.
- A **lower CV** indicates more stability or consistency, while a **higher CV** indicates more variability.

---

### **Example**

Imagine two datasets representing the weights of apples and oranges:
- **Apples**: Mean = 150g, SD = 10g
- **Oranges**: Mean = 200g, SD = 20g

**Coefficient of Variation**:
- Apples: \($ \text{CV} = \frac{10}{150} \times 100 = 6.67\% \$)

This shows that apples are more consistent in weight compared to oranges.

### **What is a Z-Score?**

A **Z-score** measures how many standard deviations a data point is from the mean of the dataset. It helps standardize data and compare values across different datasets with varying scales or units.

---

### **Z-Score Formula**

\$[
Z = \frac{(X - \mu)}{\sigma}
\]$

Where:
- \($ Z $\): Z-score
- \($ X $\): Data point
- \($ \mu $\): Mean of the dataset
- \( $\sigma $\): Standard deviation of the dataset

---

### **Why Is Z-Score Useful?**
1. Identifies how unusual a data point is compared to the rest of the dataset.
2. Helps detect outliers (e.g., Z-scores beyond Â±3 are often considered outliers).
3. Allows comparisons across datasets by converting them to a standard scale.

---

### **Simple Example**

Consider a dataset: **[10, 12, 14, 16, 18]**

#### **Step-by-Step Calculation**
1. **Calculate the Mean (\( \mu \))**:
   \[$
   \mu = \frac{10 + 12 + 14 + 16 + 18}{5} = 14
   \]$

2. **Calculate the Standard Deviation (\( \sigma \))**:
   - \( $\sigma = \sqrt{\frac{\sum (X - \mu)^2}{N}} $\)
   - Variance \($ = \frac{(10-14)^2 + (12-14)^2 + (14-14)^2 + (16-14)^2 + (18-14)^2}{5} = 8 $\)
   - Standard deviation \($ \sigma = \sqrt{8} = 2.83$ \)

3. **Calculate Z-Scores**:
   \[$
   Z = \frac{X - \mu}{\sigma}
   \]$
   - For \($ X = 10 \): \( Z = \frac{10 - 14}{2.83} = -1.41 \)$
   - For \($ X = 12 \): \( Z = \frac{12 - 14}{2.83} = -0.71 \)$
   - For \($ X = 14 \): \( Z = \frac{14 - 14}{2.83} = 0 \)$
   - For \($ X = 16 \): \( Z = \frac{16 - 14}{2.83} = 0.71 \)$
   - For \($ X = 18 \): \( Z = \frac{18 - 14}{2.83} = 1.41 \)$

---





### **Key Insights**
1. **Z-Score Interpretation**:
   - **Negative Z-Score**: Data point is below the mean.
   - **Zero Z-Score**: Data point equals the mean.
   - **Positive Z-Score**: Data point is above the mean.

2. **Practical Application**:
   - A Z-score helps identify unusual values or outliers in a dataset.
   - For example, Z-scores beyond Â±3 are considered extreme in a normal distribution.


### Understanding Z-Score with Simple Terms

The **z-score** tells us how far a value is from the average in terms of the spread of the data. It helps standardize data, so we can compare values from different scales.

The formula is:

$
z = \frac{\text{Value} - \text{Average}}{\text{Spread}}
$

Where:
- **Value** is the data point youâ€™re analyzing.
- **Average** is the mean of the data.
- **Spread** is the standard deviation, which measures how spread out the data is.

---

#### Step 2: Calculate Z-Scores for Each Value

  $

### **Different Types of Data**

Data can be classified into distinct types based on their nature, characteristics, and measurement scales. Understanding data types is crucial for selecting appropriate statistical methods and visualization techniques.

---

### **1. Types of Data Based on Nature**

#### **1.1. Qualitative Data (Categorical Data)**
- Represents categories, labels, or groups.
- Non-numerical but can sometimes be assigned numerical codes.
- **Examples**: Gender (Male, Female), Marital Status (Single, Married).

**Types of Qualitative Data**:
1. **Nominal Data**:
   - Categories with no inherent order.
   - Example: Blood Groups (A, B, AB, O).

2. **Ordinal Data**:
   - Categories with a meaningful order, but the intervals between them are not uniform.
   - Example: Education Level (High School, Bachelorâ€™s, Masterâ€™s).

---

#### **1.2. Quantitative Data (Numerical Data)**
- Represents measurable quantities.
- Always numerical in nature.
- **Examples**: Age, Salary, Height.

**Types of Quantitative Data**:
1. **Discrete Data**:
   - Counts or whole numbers.
   - Example: Number of students in a class.

2. **Continuous Data**:
   - Measured values that can take any value within a range.
   - Example: Temperature, Weight.

---

### **2. Types of Data Based on Measurement Levels**

#### **2.1. Nominal Scale**
- Represents data that can only be categorized.
- No ordering or ranking.
- **Example**: Eye Color (Blue, Green, Brown).

#### **2.2. Ordinal Scale**
- Represents data with a meaningful order but unequal intervals.
- **Example**: Movie Ratings (1 star, 2 stars, 3 stars).

#### **2.3. Interval Scale**
- Numerical data with equal intervals between values but no true zero.
- **Example**: Temperature in Celsius (0Â°C does not indicate the absence of temperature).

#### **2.4. Ratio Scale**
- Numerical data with equal intervals and a true zero, allowing for meaningful ratios.
- **Example**: Height, Weight (e.g., 60 kg is twice as heavy as 30 kg).

# **Univariate Analysis**

**Univariate analysis** is the simplest form of data analysis, focusing on analyzing a single variable (or feature) at a time. The goal is to understand the distribution, central tendency, and dispersion of the data for that variable.

---

### **Key Components of Univariate Analysis**

1. **Types of Variables**:
   - **Numerical (Continuous/Discrete)**: Variables with numeric values (e.g., Age, Salary).
   - **Categorical**: Variables with distinct categories (e.g., Gender, Marital Status).

2. **Analysis Methods**:
   - **Numerical Variables**:
     - Summary statistics: Mean, Median, Mode, Variance, Standard Deviation.
     - Visualization: Histograms, Box Plots, KDE Plots.
   - **Categorical Variables**:
     - Frequency counts and proportions.
     - Visualization: Bar Charts, Pie Charts.

---

### **Steps in Univariate Analysis**

#### **1. For Numerical Variables**
- Check summary statistics (mean, median, standard deviation).
- Visualize the distribution using histograms or box plots.
- Identify outliers or skewness.

#### **2. For Categorical Variables**
- Count the frequency of each category.
- Visualize the proportions using bar charts or pie charts.
